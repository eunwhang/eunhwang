#Topic: Generating a Wordcloud image

##Step 1: Prepare your text

```python
text = "want to see a panda up close. Time is running out. by early nest year, there won't be any of the bears left in zoos in the united states. they are being sent back to china. from more than 50 years the country has loaned pandas to American zoos as a token of good will. but for now, that period is coming to an end. we are going to miss these bears so much. Trish Jarvis tole time for kids. she is an animal keeper at the smithsonian's national zoo, in washington D.C. the zoo is sending its pandas back to china in the coming weeks. pandas have been part of everyone's lives for over 50 years. the china wildlife and conservation association owns the bears. its loan agreement with the smithsonian expires in december. some observers say china may be taking its pandas back from the us. because of disagreements between the countries' governments over issuessuch as trade and technology."
```

##Step 2: Run the following code
Think about which parts you need to change or revise

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud
wordcloud = WordCloud(width = 800, height = 600,
                      background_color ='white',
                      min_font_size = 10).generate(text)

# Display the word cloud using matplotlib
plt.imshow(wordcloud)
plt.show()
```

##(Modified) Margin, padding
Padding in image generation refers to the process of adding extra pixels around the edges of an image.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate word cloud
wordcloud = WordCloud(width = 800, height = 600,
                      background_color ='white',
                      min_font_size = 10).generate(text)

# Display the word cloud using matplotlib
# plt.figure(figsize = (8, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 2) # margine space

plt.show()
```

##(Modified 2) Font color to change
Hexa color

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Function to change the font colors
def gradient_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    shades_of_blue = ["#c41454", "#32a879", "#32a879", "#c2c73a", "#351182"]
    return shades_of_blue[len(word) % len(shades_of_blue)]


# Generate word cloud
wordcloud = WordCloud(width = 800, height = 600,
                      background_color ='white',
                      min_font_size = 10,
                      color_func=gradient_color_func  # Use the color function
                     ).generate(text)

# Display the word cloud using matplotlib
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 2)  # margin space

plt.show()
```
